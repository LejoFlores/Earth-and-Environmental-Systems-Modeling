```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
Nt = 200
Nx = 1000

mu_x = 1.5
s_x = 2.1
min_x = 1.25

# Initialize an array to hold the final positions of each walker
final_positions = np.zeros(Nx)

# Loop over each walker
for i in range(Nx):
    # Start at position 0
    position = 0
    
    # Loop over each step
    for t in range(Nt):
        # Generate a step size that is greater than min_x
        step_size = np.random.normal(mu_x, s_x)
        while step_size <= min_x:
            step_size = np.random.normal(mu_x, s_x)
        
        # Decide the direction of the step (left or right)
        step_direction = np.random.choice([-1, 1])
        
        # Update the position
        position += step_direction * step_size
    
    # Store the final position of this walker
    final_positions[i] = position

# Plotting Figure 1: Histogram of Final Positions
plt.figure(figsize=(12, 6))
plt.hist(final_positions, bins=30, range=(-100, 100), alpha=0.7, edgecolor='black')
plt.title('Histogram of Final Positions')
plt.xlabel('Final Position')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()

# Plotting Figure 2: Histogram of Step Sizes
# To obtain step sizes, we need to gather all step sizes during the walks
all_step_sizes = []

# Recompute all step sizes for the histogram
for i in range(Nx):
    position = 0
    for t in range(Nt):
        step_size = np.random.normal(mu_x, s_x)
        while step_size <= min_x:
            step_size = np.random.normal(mu_x, s_x)
        all_step_sizes.append(step_size)

plt.figure(figsize=(12, 6))
plt.hist(all_step_sizes, bins=30, range=(min_x, 10*max(all_step_sizes)), alpha=0.7, edgecolor='black')
plt.title('Histogram of Step Sizes')
plt.xlabel('Step Size')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
Nt = 300
Nx = 3

mu_x = 0.5
s_x = 1.5
mu_y = 5.0
s_y = 2.5

# Initialize arrays to store positions
positions_x = np.zeros((Nx, Nt+1))
positions_y = np.zeros((Nx, Nt+1))

# Perform random walks for each walker
for i in range(Nx):
    x_position = 0
    y_position = 0
    
    # Store the starting position
    positions_x[i, 0] = x_position
    positions_y[i, 0] = y_position
    
    # Generate each step
    for t in range(1, Nt+1):
        # Generate step sizes
        step_x = np.random.normal(mu_x, s_x)
        step_y = np.random.normal(mu_y, s_y)
        
        # Update positions
        x_position += step_x
        y_position += step_y
        
        # Store the position at this step
        positions_x[i, t] = x_position
        positions_y[i, t] = y_position

# Plotting the trajectories of all walkers
plt.figure(figsize=(12, 8))

# Plot each walker's trajectory
for i in range(Nx):
    plt.plot(positions_x[i], positions_y[i], label=f'Walker {i+1}')

# Adding labels and title
plt.title('2D Random Walks of 3 Walkers')
plt.xlabel('X Position')
plt.ylabel('Y Position')
plt.legend()
plt.grid(True)
plt.show()
```
